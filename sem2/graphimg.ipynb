```python
import sys
import pickle
import networkx as nx
from vispy import app, scene
from vispy.visuals.graphs import layouts

# --- Force a working backend ---
app.use_app('pyqt5')  # or 'pyqt5' if you have PyQt5 installed



adj_mat = nx.adjacency_matrix(G)

canvas = scene.SceneCanvas(title='NetworkX Graph from Pickle', size=(600, 600),
                           bgcolor='white', show=True)
view = canvas.central_widget.add_view('panzoom')

layout = layouts.get_layout('force_directed', iterations=100)

visual = scene.visuals.Graph(
    adj_mat, layout=layout, line_color='black', arrow_type="stealth",
    arrow_size=30, node_symbol="disc", node_size=20,
    face_color=(1, 0, 0, 0.2), border_width=0.0, animate=True,
    directed=G.is_directed(),
    parent=view.scene
)

@canvas.events.draw.connect
def on_draw(event):
    if not visual.animate_layout():
        canvas.update()

if __name__ == '__main__':
    if sys.flags.interactive != 1:
        app.run()
```
